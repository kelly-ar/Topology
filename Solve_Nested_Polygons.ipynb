```python
# -*- coding: utf-8 -*-
"""
Generated by ArcGIS ModelBuilder on : 2022-12-05 12:11:23
"""
import arcpy


def Solve_Nested_Polygons():

    # To allow overwriting outputs change overwriteOutput option to True.
    arcpy.env.overwriteOutput = False


    # Process: Erase (Erase) (analysis)
    nestedPolyErased = fr"{scratch}\nestedPoly_erased"
    arcpy.analysis.Erase(in_features=Input_Boundary, erase_features=nested_polygons, 
                         out_feature_class=nestedPolyErased, cluster_tolerance="")

    # Process: Update (Update) (analysis)
    arcpy.analysis.Update(in_features=nestedPolyErased, update_features=nested_polygons, 
        out_feature_class=Output_Name, keep_borders="BORDERS", cluster_tolerance="")

if __name__ == '__main__':
    Solve_Nested_Polygons()
```
